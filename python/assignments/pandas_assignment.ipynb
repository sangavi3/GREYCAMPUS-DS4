```python
avg_time = df['Absenteeism Time in Hours'].groupby(by = df['Reason for Absence']).mean()
avg_time
```

```python
unique_reason = sorted(df['Reason for Absence'].unique())
print(unique_reason)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,10))
plt.plot(unique_reason,avg_time)
plt.title("Line Graph between Reason for absence and Avg duration of absence")
plt.xlabel("Reason for Absence")
plt.ylabel("Avg. Absenteeism Time in Hrs")
plt.xticks(np.arange(0,30))
plt.show()
#This shows that for reason 9, average time relatively more high than any of the reasons.
# Also, for reason 0, abg time is 0
```

```python
plt.scatter(df['Distance to Work'],df['Absenteeism Time in Hours'])
plt.xlabel("Distance to Work")
plt.ylabel("Absenteeism Time in Hrs")
plt.show()
#There is almost no relationship between distance to work and absenteeism time in hrs.
```
